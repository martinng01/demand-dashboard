```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Linear Trend

```python
# Generate for 120 months (10 years)

datapoints = 120

linear_trend = np.arange(0, datapoints)
f0 = 0.6*linear_trend + 200

plt.figure(figsize=(15, 5))
plt.plot(f0)
```

# Seasonality

```python
halfyear_dist = (linear_trend % 12) - 6
halfyear_dist_norm = halfyear_dist * np.pi / 6
seasonality = np.sin(halfyear_dist_norm)
plt.plot(seasonality)
```

```python
f1 = f0 + seasonality * linear_trend/10
plt.figure(figsize=(15, 5))
plt.plot(f1)
```

# Noise

```python
noise = np.random.normal(0, 4, datapoints)

f2 = f1 + noise

plt.figure(figsize=(15, 5))
plt.plot(f2)
```

# Convert to dataframe

```python
df = pd.DataFrame(f2, columns=['units_sold'])
df = df.round(2)
df.index = pd.date_range(start='2014-01-01', end='2023-12-01', freq='MS')   

df.head()
```

```python
df.plot(figsize=(15, 5))
```

# Write to CSV

```python
df.to_csv('milk_consumption.csv')
```
